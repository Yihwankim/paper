```python
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pandas as pd
```

```python
df = pd.read_excel('2017_2Q_data.xlsx', header=0, skipfooter=0)
```

```python
#pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
#pd.set_option('display.width', 1000)
```

```python
df
```

```python
df1 = df[['거래금액', 'AGE', '세대수', '주차대수_세대',  '용적률',  '건폐율', '전용면적(㎡)', '전용률(%)', '방 개수', '화장실 개수', '층', 'D2', 'D3', 'D5', 'D6']]
```

```python
model = smf.ols(formula = '거래금액 ~ AGE + 세대수 + 주차대수_세대 + 용적률 + 건폐율 + 전용면적(㎡) + 전용률(%) + 방 개수 + 화장실 개수 + 층 + D2 + D3 + D5 + D6', data = df1)
result = model.fit()
result.summary()
```
